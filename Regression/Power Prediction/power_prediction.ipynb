# preprocessing

```python
data.drop("Date/Time",axis=1,inplace=True)
```

```python
df.drop("Date/Time",axis=1,inplace=True)
```

```python
df
```

```python
from sklearn import preprocessing
labelencoder=preprocessing.LabelEncoder()
data['City']=labelencoder.fit_transform(data['City'])
data['Country']=labelencoder.fit_transform(data['Country'])
df['City']=labelencoder.fit_transform(df['City'])
df['Country']=labelencoder.fit_transform(df['Country'])
df
```

```python
from sklearn import preprocessing
data_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
data_scale= data_scaler.fit_transform(data.values)
x= data_scaler.fit_transform(df)
y=data_scale[:,0]
```

# spliting data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model1=LinearRegression()
model1.fit(x_train,y_train)
y_pred = model1.predict(x_test)
```

```python
model1.score(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error
error=mean_absolute_error(y_test,y_pred)
print(error)
```

```python
from sklearn.tree import DecisionTreeRegressor
model1=DecisionTreeRegressor(max_depth=7,random_state=0)
model1.fit(x_train,y_train)
y_pred = model1.predict(x_test)
```

```python
model1.score(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error
error=mean_absolute_error(y_test,y_pred)
print(error)
```

```python
from sklearn.ensemble import RandomForestRegressor
model1=RandomForestRegressor(n_estimators=15,max_depth=7,random_state=0)
model1.fit(x_train,y_train)
y_pred = model1.predict(x_test)
```

```python
model1.score(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error
error=mean_absolute_error(y_test,y_pred)
print(error)
```

```python
from sklearn.neighbors import KNeighborsRegressor
model1=KNeighborsRegressor()
model1.fit(x_train,y_train)
y_pred = model1.predict(x_test)
```

```python
model1.score(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error
error=mean_absolute_error(y_test,y_pred)
print(error)
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor
model1= HistGradientBoostingRegressor()
model1.fit(x_train,y_train)
y_pred=model1.predict(x_test)
```

```python
model1.score(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error
error=mean_absolute_error(y_test,y_pred)
print(error)
```
